```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from time import time
```

```python
import fourier
```

```python
img = np.array(Image.open('../../images/beatles.png').convert('L'))

plt.gray()
plt.imshow(img)
plt.show()
```

```python
transform = fourier.Fourier()
transform.transform2D(img)
```

```python
plt.figure()
fourier.plot3D(transform.real2D,img.shape[0],img.shape[1])
plt.figure()
fourier.plot3D(transform.imag2D,img.shape[0],img.shape[1])
```

```python
# Generate a normal distribution and take inverse Fourier transform
F = np.random.normal(128,20,(200,200))
transform = fourier.Fourier()
transform.transform2D(F)

fourier.plot2D(transform.real2D,200,200)
```
