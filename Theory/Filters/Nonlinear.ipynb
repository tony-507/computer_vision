```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from time import time
```

In this notebook, we try to implement nonlinear filters as well as some special filtering techniques.

```python
import filters
import components
```

```python
# Import an image
img = np.array(Image.open('../../images/beatles.png').convert('L'))

plt.gray()
plt.imshow(img)
plt.show()
```

```python
# Median filter
t = time()
img_med = filters.median(img)
print(f'{time()-t}s used')

plt.gray()
plt.imshow(img_med)
plt.show()
```

```python
# Convert to binary image
img_bin = (img >= 128)*1

plt.gray()
plt.imshow(img_bin)
plt.show()
```

```python
# Dilation
t = time()
img_dilate = filters.dilate(img_bin)
print(f'{time()-t}s used')

plt.gray()
plt.imshow(img_dilate)
plt.show()
```

```python
# Erosion
t = time()
img_erode = filters.erode(img_bin)
print(f'{time()-t}s used')

plt.gray()
plt.imshow(img_erode)
plt.show()
```

```python
# Manhattan distance transform of binary image
t = time()
img_dist_transform = filters.block_dist_transform(img_bin)
print(f'{time()-t}s used')

plt.gray()
plt.imshow(img_dist_transform)
plt.show()
```

```python
# A simpler binary image for segmentation using component's connectivity
img = np.array(Image.open('../../images/binary_test.jpg').convert('L'))
img = (img > 128)*1

plt.gray()
plt.imshow(img)
plt.show()
```

```python
# Finding components in the image
components = components.Components(img)
components.find_components()
components.plot_components()
```
